# TRAINING A LOGISTIC REGRESSION MODEL USING THE TRAINING DATASET TO PREDICT CUSTOMER CHURN.


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Convert categorical variables into numerical
df = pd.get_dummies(df, drop_first=True)
```

```python
# Split data into features and target
X = df.drop(columns=['Exited'])  # Adjust if the target column name is different
y = df['Exited']
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)
```

```python
# Standardize features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Train logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(X_test)
```

```python
# Evaluate the model
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print("Classification Report:\n", classification_report(y_test, y_pred))
```

# Check for Overfitting

```python
train_accuracy = model.score(X_train, y_train)
test_accuracy = model.score(X_test, y_test)

print(f"Training Accuracy: {train_accuracy:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")
```

***The model's training accuracy (0.9992) and test accuracy (0.9980) are very close, indicating that no significant overfitting is present. The slight difference is expected and suggests the model is generalizing well to unseen data.


#  Analyzing the Confusion Matrix


```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay


y_pred = model.predict(X_test)

cm = confusion_matrix(y_test, y_pred)  # Compute the confusion matrix

disp = ConfusionMatrixDisplay(confusion_matrix=cm)# Display the confusion matrix
disp.plot(cmap="Blues")
plt.show()

print("Confusion Matrix:") # Print confusion matrix values
print(cm)
```

```python
TN, FP, FN, TP = cm.ravel()

accuracy = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP) if (TP + FP) != 0 else 0
recall = TP / (TP + FN) if (TP + FN) != 0 else 0
f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) != 0 else 0

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1_score:.4f}")

```

---The model is highly accurate with minimal errors

# Evaluate with ROC Curve & AUC Score

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve,roc_auc_score
```

```python
# Compute ROC curve
fpr, tpr, _ = roc_curve(y_test, model.predict_proba(X_test)[:, 1])

# Calculate AUC
auc = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1]) # Calculate and assign auc

# Plot ROC curve
plt.plot(fpr, tpr, label=f'AUC = {auc:.4f}')
plt.plot([0, 1], [0, 1], 'r--')  # Random classifier line
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

The model is highly accurate and reliable, with minimal misclassifications. No major improvements are needed unless reducing false positives or negatives is critical.

#Check for Overfitting


```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Train Random Forest Model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predict on Training and Test Data
train_pred = rf_model.predict(X_train)
test_pred = rf_model.predict(X_test)

# Calculate Accuracy
train_acc = accuracy_score(y_train, train_pred)
test_acc = accuracy_score(y_test, test_pred)

print(f"Training Accuracy: {train_acc:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")

# Check Overfitting
if train_acc - test_acc > 0.05:
    print("âš ï¸ Possible Overfitting Detected!")
else:
    print("âœ… No major overfitting.")
```

# Evaluate with ROC Curve & AUC Score

```python
from sklearn.metrics import roc_curve, roc_auc_score

# Get probability scores for ROC Curve
y_rf_prob = rf_model.predict_proba(X_test)[:, 1]

# Compute ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_rf_prob)
auc_score = roc_auc_score(y_test, y_rf_prob)

# Plot ROC Curve
plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f"Random Forest (AUC = {auc_score:.4f})", color='blue')
plt.plot([0, 1], [0, 1], 'r--', label="Random Classifier")  # Baseline
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Random Forest")
plt.legend()
plt.show()

print(f"AUC Score: {auc_score:.4f}")
```

# Evaluate Random Forest with K-Fold Cross-Validation

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier


# Initialize Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Perform K-Fold Cross-Validation (K=10)
k = 10
cv_scores = cross_val_score(rf_model, X, y, cv=k, scoring='accuracy')

# Print results
print(f"K-Fold Cross-Validation Results ({k}-folds):")
print(f"Mean Accuracy: {np.mean(cv_scores):.4f}")
print(f"Standard Deviation: {np.std(cv_scores):.4f}")
```
